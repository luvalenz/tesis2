```python
import dill
import numpy as np
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode()
```

```python
with open('/home/lucas/tesis2/test_outputs/old/sequence_tree_1000samples_20levels_100samples_per_class_20results_per_query.dill', 'rb') as f:
    test_results_1000 = dill.load(f)
with open('/home/lucas/tesis2/test_outputs/old/sequence_tree_5000samples_20levels_100samples_per_class_20results_per_query.dill', 'rb') as f:
    test_results_5000 = dill.load(f)
with open('/home/lucas/tesis2/test_outputs/old/sequence_tree_20000samples_20levels_100samples_per_class_20results_per_query.dill', 'rb') as f:
    test_results_20000 = dill.load(f)
```

```python
def plot_times(results_dict, title):
    data = []
    for model_name, result in sorted(results_dict.items()):
        x = []
        y = []
        for class_name, scores in sorted(result.items()):
            x += len(scores)*[class_name]
            y += scores
        trace = go.Box(
            y=y,
            x=x,
            name=model_name,
            boxmean=True
        )
        data.append(trace)
    layout = go.Layout(
        yaxis=dict(
            title=title,
            zeroline=True
        ),
        boxmode='group'
    )
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)
```

```python
models = [test_results_1000, test_results_5000, test_results_20000]
models_times_by_class = [model['times'] for model in models]
models_names = [1000, 5000, 20000]
models_times = {name: times for name, times in zip(models_names, models_times_by_class)}
plot_times(models_times, 'seconds')
```
